## Luftspektren at Different Resolutions

```python
# Loop over directory

import os

directory = os.fsencode("/Users/lukas/Desktop/Studium/Chemie/Physikalische_Chemie/PC_Praktikum/QMS/Raw Data")
directory_string = os.fsdecode(directory)



data_list = []
filenames = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)

    if filename.startswith("probe") and filename.endswith("_7") and not "35" in filename:
        # Write The String filname_multiplier_resolution_empf

        string_list = filename.split("_")

        # Make String
        string = "Probe" + " " + "Res=" + string_list[2] + "%" + ", " + "Mult=30 %, Empf=$10^{-7}$"
        filenames.append(string)

        df = pd.read_csv(os.path.join(directory_string, filename), sep="\t", names=[0, "mz","intensity"])
        # Remove the first column
        df = df.iloc[:, 1:]
        # Append the DataFrame to the list
        data_list.append(df)



# Create a figure for the plots
plt.figure(figsize=(14, 6))
# Loop over the DataFrames in the list and plot each one
for i, df in enumerate(data_list):
    sns.lineplot(data=df, x="mz", y="intensity", label=filenames[i], linewidth=1.5)
plt.xlabel("m/z")
plt.ylabel("Intensity / -")
plt.legend(title="Legende", loc="upper right", fancybox=False, edgecolor="black")
# Add ticks
plt.xticks(np.arange(0, 50, 1))
plt.tight_layout()
plt.grid()
plt.savefig("variation_der_resolution.png", dpi=800, bbox_inches='tight')


```

```python
# Calculate Ratio of Intensities of all Peaks to 30 % as a Reference
intensities_30 = peaks_30_30_7_df["intensity"].values

intensities_45 = peaks_30_45_7_df["intensity"].values

intensities_60 = peaks_30_60_7_df["intensity"].values

intensities_75 = peaks_30_75_7_df["intensity"].values

# make latex table of mz of 30 and all the intensities

latex_table = pd.DataFrame(
    {"m/z": peaks_30_30_7_df["mz"].values,
     "Intensity_30": intensities_30,
     "Intensity_45": intensities_45,
     "Intensity_60": intensities_60,
     "Intensity_75": intensities_75
     }
)
latex_table = latex_table.round(2)
latex_table = latex_table.to_latex(index=False, float_format="%.2f", escape=False)
print(latex_table)

# Make a DataFrame with the ratios and m/z of and all ratios

ratios = pd.DataFrame(
    {"m/z": peaks_30_30_7_df["mz"].values,
     "Ratio_30": intensities_30/intensities_30 * 100,
     "Ratio_45": intensities_45/intensities_30 * 100 ,
     "Ratio_60": intensities_60/intensities_30 * 100,
     "Ratio_75": intensities_75/intensities_30 * 100
     }
)

# Make a plot of all the ratios

plt.figure(figsize=(12, 6))
sns.lineplot(data=ratios, x="m/z", y="Ratio_45", label="Ratio: Res 45 % / 30 %", color="darkred", linewidth=1.5)
sns.lineplot(data=ratios, x="m/z", y="Ratio_60", label="Ratio: Res 60 % / 30 %", color="darkgreen", linewidth=1.5)
sns.lineplot(data=ratios, x="m/z", y="Ratio_75", label="Ratio: Res 75 % / 30 %", color="darkorange", linewidth=1.5)
plt.xlabel("m/z")
plt.ylabel("Relative Intensität zu Res 30 % / %")
plt.xticks(np.arange(0, 50, 1))

plt.legend(title="Legende", loc="upper right",fancybox=False, edgecolor="black")
plt.grid()
plt.tight_layout()
plt.savefig("peak_heights_ratio.png", dpi=800, bbox_inches='tight')
plt.show()
```
