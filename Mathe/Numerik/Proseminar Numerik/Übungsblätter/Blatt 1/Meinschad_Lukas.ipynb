## Teilaufgabe c

Es wurden einige Daten gefunden

nun muss man in doppelt und einfach logarithmischer Skalierung auftragen und dann eine Funktion finden



```python
x_dat = [2,3,4,5,7.5,10,16]
y_dat = [3.53553, 2.88675, 2.5, 2.23607, 1.82574, 1.58114, 1.25]

x_dat_log = [np.log(x) for x in x_dat]
y_dat_log = [np.log(y) for y in y_dat]

fig, axs = plt.subplots(1, 3, figsize=(18, 5))


# First subplot
axs[0].plot(x_dat, y_dat, label='normal data', color='blue')
axs[0].set_title('Normal data')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend()

# Second subplot
axs[1].plot(x_dat, y_dat_log, label='log(data)', color='blue')
axs[1].set_title('log data against x')
axs[1].set_xlabel('x')
axs[1].set_ylabel('log(data)')
axs[1].legend()

# Third subplot
axs[2].plot(x_dat_log, y_dat_log, label='log(data) vs log(x)', color='blue')
axs[2].set_title('log(data)  vs log(x)')
axs[2].set_xlabel('log(x)')
axs[2].set_ylabel('log(data)')
axs[2].legend()

plt.tight_layout()
plt.show()

# man sieht hier klar in doppelt logarithmischer Auftragung hat man linearisierung
# damit als Fit funktion
```

```python
from scipy.optimize import curve_fit

# Wir wissen also wir haben es mit Potenzfunktion zutun 
def model(x, C, k):
    return C * x**k

# Führe den Fit durch
params, covariance = curve_fit(model, x_dat, y_dat)

# Extrahiere die Parameter C und k
C_fit, k_fit = params

print(C_fit,k_fit)

# Berechne die angepasste Funktion
y_fit = model(x_dat, C_fit, k_fit)

# Plot der Daten und des Fits
plt.figure(figsize=(8, 6))
plt.plot(x_dat, y_dat, 'o', label='Messwerte')
plt.plot(x_dat, y_fit, '-', label=f'Fit: y = {C_fit:.2f} * x^{k_fit:.2f}', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.title('Fit der Funktion y = C * x^k')
plt.show()

C_fit, k_fit
```

# Aufgabe 3

Es sei für $x_0 >0$ die folgende Folge gegeben rekursiv

$x_{n+1} := \frac{1}{2}(x_n + \frac{2}{x_n})$

die folge konvergiert gegen $\sqrt{2}$

## Teilaufgabe a)


Schreiben Sie eine Funktion, welche bei Übergabe von x0 > 0 und N ∈ N die ersten N
Folgenglieder berechnet und zeichnet


```python
def calculate_and_plot_sequence(x0, N):
    # Initialisierung der Folge
    x_values = [x0]

    # Berechnung von N Folgeglieder x0 ist das erste Folgelgied
    for n in range(1, N):
        x_next = 0.5 * (x_values[-1] + 2 / x_values[-1])
        x_values.append(x_next)

    # Mit den Fertigen X Values kann man gleich in der Funktion plotten
    # Range(N) = 0 ----> N

    plt.figure(figsize=(10, 5))
    plt.plot(range(N), x_values, marker='o', label='Folge (x_n)')
    # zudem können wir die Wurzel aus 2 noch in den Plot integrieren
    plt.axhline(y=np.sqrt(2), color='r', linestyle='--', label='√2')
    plt.title('Rekursive Folge')
    plt.xlabel('n')
    plt.ylabel('x_n')
    plt.legend()
    plt.grid(True)
    plt.show()

    return x_values

x_0 =20
N = 50

calculate_and_plot_sequence(x_0,N)

```

## Teilaufgabe 2

Schreiben sie eine weitere Funktion, welche bei der Übergabe von $x_0>0$ und $\epsilon>0$ den ersten Index $n_0$ findet sodass $|x_{n0} - \sqrt(2)| < \epsilon$ ist lassen sie den $x_0$ zurückgeben

```python
# Definieren der Funktion find index
# Prinzip diese erhöht den Iterator so lange bis für ein gegebenes epsilon
# Die Bedinung value >= epsilon nicht mehr true ist
# damit hat man dann den gesuchten Folgenindex für den Epsilon "Schlauch"

def find_index(x0, epsilon):
    x_n = x0
    n = 0

    while abs(x_n - np.sqrt(2)) >= epsilon:
        x_n = 0.5 * (x_n + 2 / x_n)
        n += 1

    return n

x_0 = 5
epsilon=0.00000001

find_index(x_0,epsilon)
```
