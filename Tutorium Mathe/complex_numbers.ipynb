```python
import matplotlib.pyplot as plt

# Define complex numbers
complex_numbers = [3 + 4j]


# Extract real and imaginary parts
real_parts = [z.real for z in complex_numbers]
imaginary_parts = [z.imag for z in complex_numbers]

# Create the plot
plt.figure(figsize=(6, 6))
plt.scatter(real_parts, imaginary_parts, color='blue', label='Complex Numbers')

# Annotate each point with its corresponding complex number
for i, z in enumerate(complex_numbers):
    plt.text(real_parts[i], imaginary_parts[i], f'{z}', fontsize=12, ha='right')

# Labels and title
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Plot of Complex Numbers')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
