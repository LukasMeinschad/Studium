```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x, y) = (x + y)^2
def f(x, y):
    return (x + y)**2

# Define the constraint g(x, y) = x^2 + y^2 - 1 = 0 (unit circle)
def g(x, y):
    return x**2 + y**2 - 1

# Generate grid points for plotting
x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the contour lines of f(x, y)
plt.figure(figsize=(10, 8))
contours = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contours, label='f(x, y) = (x + y)^2')

# Plot the constraint (unit circle)
theta = np.linspace(0, 2*np.pi, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
plt.plot(circle_x, circle_y, 'r-', label='Constraint: x² + y² = 1')

# Mark the critical points (minima and maxima)
minima = [(np.sqrt(2)/2, -np.sqrt(2)/2), (-np.sqrt(2)/2, np.sqrt(2)/2)]
maxima = [(np.sqrt(2)/2, np.sqrt(2)/2), (-np.sqrt(2)/2, -np.sqrt(2)/2)]

for point in minima:
    plt.scatter(*point, color='blue', s=100, label='Minima (f=0)', zorder=5)
for point in maxima:
    plt.scatter(*point, color='red', s=100, label='Maxima (f=2)', zorder=5)

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Minimization of f(x, y) = (x + y)^2 subject to x² + y² = 1')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function f(x, y) = (x + y)^2
def f(x, y):
    return (x + y)**2

# Generate grid points for plotting
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create the 3D plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of f(x, y)
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7, edgecolor='none')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5, label='f(x, y) = (x + y)^2')

# Plot the constraint (unit circle) on the xy-plane and along the surface
theta = np.linspace(0, 2*np.pi, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
circle_z = f(circle_x, circle_y)  # Evaluate f on the circle

# Plot the unit circle in 3D (on the xy-plane)
ax.plot(circle_x, circle_y, np.zeros_like(circle_x), 'r-', linewidth=2, label='Constraint: x² + y² = 1 (xy-plane)')

# Plot the unit circle lifted onto the surface f(x, y)
ax.plot(circle_x, circle_y, circle_z, 'b-', linewidth=2, label='f(x, y) along constraint')

# Mark the critical points (minima and maxima)
minima = [(np.sqrt(2)/2, -np.sqrt(2)/2), (-np.sqrt(2)/2, np.sqrt(2)/2)]
maxima = [(np.sqrt(2)/2, np.sqrt(2)/2), (-np.sqrt(2)/2, -np.sqrt(2)/2)]

for (px, py) in minima:
    pz = f(px, py)
    ax.scatter(px, py, pz, color='blue', s=100, depthshade=False, label='Minima (f=0)' if px > 0 else "")
    
for (px, py) in maxima:
    pz = f(px, py)
    ax.scatter(px, py, pz, color='red', s=100, depthshade=False, label='Maxima (f=2)' if px > 0 else "")

# Add labels and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Plot of f(x, y) = (x + y)^2 with Constraint x² + y² = 1')
ax.legend()
plt.tight_layout()
plt.show()
```
